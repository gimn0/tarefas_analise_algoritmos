```python
import matplotlib.pyplot as plt
from statistics import mean
```

```python
entries_size = [5, 10, 15, 20, 25]

t_five_entries = [0.0004, 0.0004, 0.0004, 0.0003, 0.0003]           #
t_ten_entries = [0.0008, 0.0007, 0.0008, 0.0007, 0.0007]            #
t_fifteen_entries = [0.0016, 0.0017, 0.0016, 0.0017, 0.0016]        # arrays com os tempos de execução
t_twenty_entries = [0.0026, 0.0021, 0.0025, 0.0027, 0.002]          #
t_twenty_five_entries = [0.0037, 0.0036, 0.0029, 0.0028, 0.0028]    #


am_t_five_entries =  mean(t_five_entries)                # 
am_t_ten_entries = mean(t_ten_entries)                   #
am_t_fifteen_entries = mean(t_fifteen_entries)           # média aritmética dos tempos de execução
am_t_twenty_entries = mean(t_twenty_entries)             #
am_t_twenty_five_entries = mean(t_twenty_five_entries)   #

am_time = [am_t_five_entries, am_t_ten_entries, am_t_fifteen_entries,     # lista com as médias aritméticas
                am_t_twenty_entries, am_t_twenty_five_entries]            #

# print(am_time)
```

```python
fig, ax = plt.subplots()

plt.title('Tempo do Algoritmo Bubble Sort por Tamanho de Entrada')
plt.xlabel('tamanho das entradas')
plt.ylabel('tempo de execução (ms)')

plt.axis([2, 28, 0, 0.0035])

for x, y in zip(entries_size, am_time):
    ax.text(x, y - 0.00015, y)

ax.plot(entries_size, am_time, 'o--')


plt.show()
```
