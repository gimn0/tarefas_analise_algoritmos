```python
import matplotlib.pyplot as plt
from statistics import mean
```

```python
entries_sizes = [5, 10, 15, 20, 25] # tamanho do vetor usado

t_five_entries = [0.0004, 0.0003, 0.0003, 0.0003, 0.0003]           #
t_ten_entries = [0.0006, 0.0005, 0.0008, 0.0007, 0.0007]            #
t_fifteen_entries = [0.0014, 0.0011, 0.0012, 0.0012, 0.0013]        # arrays com os tempos de execução
t_twenty_entries = [0.0018, 0.0021, 0.0017, 0.002, 0.0021]          #
t_twenty_five_entries = [0.0028, 0.0029, 0.0023, 0.0028, 0.0028]    #


am_t_five_entries =  "{:.4f}".format(mean(t_five_entries))                # 
am_t_ten_entries = "{:.4f}".format(mean(t_ten_entries))                   #
am_t_fifteen_entries = "{:.4f}".format(mean(t_fifteen_entries))           # média aritmética dos tempos de execução
am_t_twenty_entries = "{:.4f}".format(mean(t_twenty_entries))             #
am_t_twenty_five_entries = "{:.4f}".format(mean(t_twenty_five_entries))   #

am_time = list(map(float, [am_t_five_entries, am_t_ten_entries, am_t_fifteen_entries,     # lista com as médias aritméticas
                am_t_twenty_entries, am_t_twenty_five_entries]))                          # formatado de volta para float

# print(am_time)
```

```python
fig, ax = plt.subplots()

plt.title('Tempo do Algoritmo Selection Sort por Tamanho de Entrada')
plt.xlabel('tamanho das entradas')
plt.ylabel('tempo de execução (ms)')

plt.axis([2, 28, 0, 0.003])

for x, y in zip(entries_sizes, am_time):
    ax.text(x, y - 0.00015, y)

ax.plot(entries_sizes, am_time, 'o--')


plt.show()
```
